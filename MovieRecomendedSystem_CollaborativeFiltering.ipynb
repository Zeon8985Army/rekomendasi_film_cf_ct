```python
import pandas as pd 
import numpy as np 

# Kita pakai SVD
from surprise import Reader, Dataset, SVD

from surprise.model_selection import cross_validate as evaluate

reader = Reader()

# Baca Data CSV nya
ratings=pd.read_csv('C:/Users/Lukas_K/JupyterFile/MovieRecomendedSystem/Data/ratings_small.csv')
```

```python
# Ratingnya paling tinggi adalah 5
ratings[ratings['userId']==1]
```

```python
# Hanya memakai UserId .MovieId, ratingnya. 
# Pendekatan = user based filtering (melihat rating ke movie dari user" lain)
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
```

```python
print(data)
```

```python
# pakai mesin pembelarajan SVD()
# Kode ini untuk mencari tahu apakah model yang dihasilkan akan baik 
# semakin kecil RMSE , akan makin baik
# Sekaligus menyesuaikan model dengan kondisi data yang ada
svd = SVD()
evaluate(svd, data, measures=['RMSE', 'MAE'],cv=5)
```

```python
# data.build_full_trainset() untuk membuat data latihnya
trainset = data.build_full_trainset()

# dari data latih kemudian dilatihkan dengan fit()
svd.fit(trainset)
```

```python
# Coba kita lihat rating ke moive-movie yang sudah diberikan userID 1
ratings[ratings['userId']==2]
```

```python
# Dari model yang sudah terbentuk maka dapat ditestkan
# User 1 akan memberikan score 2.67 ke movie 302, berdasarkan user 3
svd.predict(1, 302, 3)
```
