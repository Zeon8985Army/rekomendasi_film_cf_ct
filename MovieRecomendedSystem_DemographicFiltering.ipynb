```python
# # data di df1 pada Kolom dideklarasi ulang karena kolom "movie_id" tidak bisa di inner joinkan dengan kolom id di df2
# # inner join dengan "merge" mewajibkan dua kolom harus memiliki nama yang sama
# df1.columns = ['id','tittle','cast','crew']
# df2= df2.merge(df1,on='id')
```

```python
# Atau bisa menggunakan left join dengan cara memberi tahu PK(Primary Key) dan FK (Foreign Key)
df2= df2.merge(df1,left_on='id', right_on='movie_id')
```

```python
# Berapa rata-rata rating yang diberikan
C= df2['vote_average'].mean()
C
```

```python
# Mencari jumlah minimal suatu movie bisa dijadikan proses pembelajaran
m= df2['vote_count'].quantile(0.9)
m
```

```python
# loc adalah fungsi untuk mengambil baris yang value "vote_count" > "1838"
# Didapati ada 481 baris dengan 24 features

# loc sebetulnya kayak if else yang disederhanakan untuk mengambil baris data 
# yang sesuai dengan kondisi yang diberikan

# shape digunakan untuk mencari (banyak data) dan (banyak feature /"kolom" )

q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
# Rumus pembobotan rating dari IMDB
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Tambahin feature baru 'score' di "q_movies" yang nilainya didapat dari `weighted_rating()`
# axix 1 = Horizontal
# axis 0 = Vertikal

# Jadi akan dipakai data Horizontal yang berupa value dari 1 movie
# Yang dipakai adalah "vote_count" dan "vote_avarage"

q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
#Sort movies based on score calculated above
q_movies = q_movies.sort_values('score', ascending=False)

#Print the top 15 movies
q_movies[['title_y', 'vote_count', 'vote_average', 'score']].head(10)
```
