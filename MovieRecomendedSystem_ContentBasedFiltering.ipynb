```python
# Pakai TfIdfVectorizer dari scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

# Pakai TF-IDF Vectorizer Object. Buang semua stopword 'the', 'a' karena kurang terkait konteks
tfidf = TfidfVectorizer(stop_words='english')

# sebelum pembobotan, pastikan overview NaN diubah jadi '', karena akan error kalau NaN dibiarin
df2['overview'] = df2['overview'].fillna('')

# TF-IDF kan overview yang sudah dibetulkan overviewnya
tfidf_matrix = tfidf.fit_transform(df2['overview'])

# shape of tfidf_matrix (Jumlah data dan jumlah Feature)
tfidf_matrix.shape
```

```python
# Idenya adalah membandingkan nilai diantara tfidf.

# Import linear_kernel dari sklearn
from sklearn.metrics.pairwise import linear_kernel

# Hasil nilai cosine similarity matrix (kesamaan diantara hasil tfidf nya)
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

cosine_sim
```

```python
# Nilai kesamaan diantara tfidf pada setiap feature
print(cosine_sim[0])
print(cosine_sim[1])
```

```python
# karena contentBased melihat pada kesamaan kontent dari film. Maka perlu adanya inputan
# Inputan akan berupa pertanya mengenai salah satu film yang cenderung disukai pengguna
# Dari sini, inputan akan berupa judul yang akan dicari judul" lain yang memiliki nilai
# TFIDF yang serupa dengan judul film yang disebut

# Title akan dijadikan index sebagai fokus pencariannya
indices = pd.Series(df2.index, index=df2['title_y']).drop_duplicates()
```

```python
# Fungsi untuk menjadikan judul sebagai inputan dan mengeluarkan output film yang mirip
def get_recommendations(title, cosine_sim=cosine_sim):

    # Dari Index yang dibuat sebelumnya, dipakai untuk mencari baris data film dengan memakai judul 
    # idx akan mendapatkan lokasi index cosine_sim dari film yang dicari
    idx = indices[title]

    # Cari nilai cosine_sim dengan film" lain
    sim_scores = list(enumerate(cosine_sim[idx]))

    # urutkan berdasarkan nilai cosine_sim
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # ambil 10 teratas
    sim_scores = sim_scores[1:11]

    # pada nilai sim_score terdapat 2 kolom : id dan score
    # Ambil ID nya untuk dipakai mencari nama judul nya dibaris selanjutnya
    movie_indices = [i[0] for i in sim_scores]

    # Dari daftar ID yang sudah didapat, gunakan "iloc" untuk mencari baris data dari ID
    # Ambil  title dan id nya untuk dikembalikan
    return df2['title_y'].iloc[movie_indices]

#     # Jika ingin melihat score 
#     return sim_scores;
```

```python
get_recommendations('The Dark Knight Rises')
```

```python
# Kode dibawah untuk membuat dictionary di crew, cast, keywords,genres 
# {"id": 28, "name": "Action"} -> awalnya String murni dapat jadi Dictionary , Keren!
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)
```

```python
# Udah jadi dictionary
df2.head(1).crew[0][0]['department']
```

```python
# Karena sudah dictionary, maka sudah mudah mengakses valuenya. Pada crew akan tampil
# daftar seluruh crew, jika ada crew bagian director, maka ambil value tersebut untuk disendirikan
# Jika tidak ada direktor maka return NaN
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
# Returns the list top 3 elements or entire list; whichever is more.
# Kembalikan 3 element tertinggi dari setiap feature (cast,keyword,genres)
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        # Kalau pada setiap feature ada 3 element lebih, maka kembalikan 3 awal saja, supaya tidak kebanyakan
        if len(names) > 3:
            names = names[:3]
        return names

    # Jika tidak ada element sama sekali di feature tersebut, maka kembalikan kosongan
    return []
```

```python
# Dari rumus get_director maka ambil director dari dictionary crew
df2['director'] = df2['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
# Coba lihat hasil dari get_list dan get_director
df2[['title_y', 'cast', 'director', 'keywords', 'genres']].tail(8)
# Terlihat ada [] , ini harus dihilangkan karena bisa mempengaruhi
```

```python
# Karena ada [] , maka harus dihilangkan
def clean_data(x):
    # isinstance (x,list) -> jika x adalah list, maka True
    if isinstance(x, list):
        # Kalau dia list, maka lakukan iterasi untuk membaca tiap element
        # kalau ada element dengan " " -> ""
        # Lowercse kan semua biar rapi
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        # Kalau x bukan List ->"direktor bukan List tapi String"
        # Karena string maka langsung saja ubah " " -> ""
        # Lowercse kan semua biar rapi
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
# Apply clean_data ke semua feature
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
# Buat feature baru / kolom baru yang valuenya adalah gabungan dari feature" yang lain
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
```

```python
# Gunakan CV dari sklearn untuk membuat Matrix
# Yang divektorisasi hanya 1 kolom , tapi kombinasi dari 4 kolom acuan [genre,cast,director,keywords]
from sklearn.feature_extraction.text import CountVectorizer

# CV kan dari stop_words yang sudah dihilangkan
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
```

```python
# Dari matrix CV, gunakan cosine_similarity buat mencari CV" yang sama
# Bisa juga pakai linear_kernel
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
# jadikan title sebagai index untuk pencarian lokasi hasil CV nya
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title_y'])
```

```python
# Gunakan fungsi yang sudah dibuat diawal
get_recommendations('The Dark Knight Rises', cosine_sim2)
```
